```python
import json
from time import time
import pandas as pd
from pathlib import Path
from dpmm.pipelines import MSTPipeline
```

# Load data

```python
wine_dir = Path().parent / "wine"

df = pd.read_pickle(wine_dir / "wine.pkl.gz")
with (wine_dir / "wine_bounds.json").open("r") as f:
    domain = json.load(f)

df.head()
```

# Build pipeline

```python
pipeline = MSTPipeline(
    epsilon=1,  # Privacy budget for generate model
    proc_epsilon=0.1,  # Privacy budget for data processing 
    delta=1e-5,  # Delta Setting in (eps,  delta) differential privacy
)
```

# Fit pipeline

```python
print("Fitting MST pipeline...", end='\r')
start_time = time()
# The number of rows to generate
pipeline.fit(df, domain)
fit_time = time()

print(f"Fitting MST pipeline - Took {fit_time - start_time:.2f} seconds")
```

# Generate step
Deterministic with random state set to 2025

```python
print("Generating synthetic data - with fixed random state ...", end="\r")
# Generate synthetic data
synth_df = pipeline.generate(df.shape[0], random_state=2025)
gen_time = time()

print(f"Generating synthetic data - with fixed random state - Took {gen_time - fit_time:.2f} seconds")

display(synth_df.head())


print("Generating synthetic data - with fixed random state ...", end="\r")
# Generate synthetic data
synth_new = pipeline.generate(df.shape[0], random_state=2025)
gen_time = time()

print(f"Generating synthetic data - with fixed random state - Took {gen_time - fit_time:.2f} seconds")

display(synth_new.head())
```
